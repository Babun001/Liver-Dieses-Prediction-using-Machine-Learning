```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Liver_df = pd.read_csv('indian_liver_patient.csv')
Liver_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
Liver_df['Gender'] = l.fit_transform(Liver_df['Gender'])
Liver_df.head(10)
```

```python
Liver_df.Albumin_and_Globulin_Ratio = Liver_df.Albumin_and_Globulin_Ratio.fillna(Liver_df['Albumin_and_Globulin_Ratio'].mean())
Liver_df.isnull().sum()
```

```python
corr_matrix = Liver_df.corr()
corr_matrix['Dataset']*100
```

```python
X =  Liver_df.drop(['Dataset','Total_Protiens','Alamine_Aminotransferase','Total_Protiens','Direct_Bilirubin','Total_Bilirubin'], axis=1)
y = Liver_df['Dataset']
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test =train_test_split(X,y,test_size=0.2, random_state=1)
print(X_train.shape, X_test.shape)
X_train.shape[0]+ X_test.shape[0]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
print(X_train_sc)
print("----------------------")
print(X_test_sc)
```

```python
X_train_sc = pd.DataFrame(X_train_sc, columns=["Age","Gender","Alkaline_Phosphotase","Aspartate_Aminotransferase","Albumin","Albumin_and_Globulin_Ratio"])
X_test_sc = pd.DataFrame(X_test_sc, columns=["Age","Gender","Alkaline_Phosphotase","Aspartate_Aminotransferase","Albumin","Albumin_and_Globulin_Ratio"])
X_train_sc.describe().round()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
scores = cross_val_score(DecisionTreeClassifier(),X,y, cv=5)
scores
```

```python
scores.mean()*100
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0    
)
bag_model.fit(X_train_sc,y_train)
bag_model.oob_score_*100 #68.24
```

```python
bag_model.score(X_test_sc,y_test)*100 # 67
```

```python
bag_model = BaggingClassifier(
    base_estimator = DecisionTreeClassifier(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0    
)

scores = cross_val_score(bag_model,X,y, cv=5)
scores.mean()*100 # 68.78573533745947
```

```python
from sklearn.ensemble import RandomForestClassifier

scores = cross_val_score(RandomForestClassifier(), X,y,cv=5)
scores.mean()*100 # 69.8113763631005
```
