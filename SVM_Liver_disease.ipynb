```python
# feature droping 
X=Liver_df.drop(["Direct_Bilirubin","Aspartate_Aminotransferase","Total_Protiens", "Dataset"], axis=1)
```

```python
y=Liver_df["Dataset"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
print(len(X_train)) #466
print(len(X_test)) #117
```

```python
from sklearn.svm import SVC
Score = {}
C = [1,10,20,30]
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
gamma = ['scale', 'auto']
for c in C:
    for k in kernel:
        model = SVC(C=c,kernel=k)
        model.fit(X_train,y_train)
        acc = model.score(X_test,y_test)*100
        Score[c,k] = acc      
                
tunedVal = max(Score,key=Score.get)
tunedVal,max(Score.values())
```
